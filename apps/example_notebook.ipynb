```python
lenna = cv2.imread('Lenna.png')
lenna_rgb = cv2.cvtColor(lenna, cv2.COLOR_BGR2RGB)
plt.imshow(lenna_rgb)
```

```python
# load the pre-trained model
classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
# perform face detection
bboxes = classifier.detectMultiScale(lenna_rgb)
# print bounding box for each detected face
for box in bboxes:
    # extract
    x, y, width, height = box
    x2, y2 = x + width, y + height
    # draw a rectangle over the pixels
    cv2.rectangle(lenna_rgb, (x, y), (x2, y2),(0,255,0), 4)
    cv2.putText(lenna_rgb, "Lenna", (x+x2, y+y2), cv2.FONT_HERSHEY_TRIPLEX, 2, (0,255,0), 3)
# show the image
plt.imshow(lenna_rgb)
```
